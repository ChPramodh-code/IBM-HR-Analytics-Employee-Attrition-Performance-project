Data Analysis:

Data Analysis:

```python
# The Graph of Employees in IBM in different Age Groups:
ax = sb.histplot(data=dataset, x='Age',kde = True)
ax.set_title("Distribution of employees as per Age")
ax.set(xlabel="Age")
ax.set(ylabel="count of employees at the age group")
```

```python
# The Graph Of Male VS Female Employees working in IBM:
ab=sb.countplot(data = dataset,x='Gender')
ab.set_title("Count of Employees in Male and Female Gender")
ab.set(xlabel="gender")
ab.set(ylabel="Count of employees")
```

```python
a=dataset['Gender']
male_count = 0
female_count = 0
for i in a:
    if i == 'Male':
        male_count = male_count + 1
    if i == 'Female':
        female_count = female_count + 1
if female_count > male_count:
    print(f"female employees is greater than male employees with difference of {female_count-male_count} ")
else:
    print(f"male employees is greater than female employees with difference of {male_count-female_count} ")
```

```python
# The Graph Of Employees in different departments in IBM:
ac=sb.countplot(data = dataset,x='Department')
ac.set_title("Count of Employees in different departments")
ac.set(xlabel="departments")
ac.set(ylabel="Count of employees")
```

```python
b = dataset['Department'].value_counts()
print(b)
```

```python
# The Graph Of count of Attrition in IBM:
af=sb.countplot(data = dataset , x='Attrition')
af.set_title("The Graph of the count of Attrition in IBM")
af.set(xlabel="Attrition reponses")
af.set(ylabel="Attrition reponses count")
```

```python
attrition_a= dataset['Attrition'].value_counts()
print(attrition_a)
```

```python
attrition= dataset['Attrition'].value_counts(normalize=True)
aa = attrition* 100
print(aa)
```

```python
ad = sb.barplot(x=attrition.index, y=attrition)
ad.set_title("percentage of attrition")
ad.set(xlabel="attrition")
ad.set(ylabel="percentage")
```

```python
aa=sb.countplot(data = dataset , x='JobSatisfaction')
aa.set_title("Job Satisfaction Rating by Employees")
aa.set(xlabel="Rating of Job Satisfaction")
aa.set(ylabel="Count of rating 1 to 5")
```
